# Problem 739

```python
def lucasSequence(n):
  x = [0]*n
  x[0]=1
  x[1]=3
  n=n-2
  for i in range(0,n):
    x[i+2] = sum(x[i:i+2])
  #print(x)
  return(x)
```

```python
def partialSummation(sequence):
  n = len(sequence)
  if n==1:
    #print(n)
    return sequence
  else:
    #print(n)
    return partialSummation( np.cumsum(sequence[1:]) )
```

```python
partialSummation( lucasSequence(10**8) ) % 1000000007
```
